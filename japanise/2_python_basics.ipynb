## 💡 Google Colabとは？

Google Colab（コラボ）は、**ブラウザだけでPythonを使える無料のツール**です。  
パソコンにPythonをインストールしなくても、すぐにコードを書いて実行できます。

- Googleのクラウド上で動く「仮想のパソコン」を使います。
- Googleアカウントがあれば、誰でも利用できます。
- コードと説明文をまとめて書けるので、学習や実験ノートにも便利です。

<br>

---

## 🧱 「セル」とは？

Google Colab のノートは、「セル」と呼ばれる**ブロック単位**でできています。

- 📄 **テキストセル**：説明や見出しを書くところ（この文もテキストセル）
- 🧮 **コードセル**：Python の命令を書くところ（ここに `print("Hello")` などを書く）

👇 セルを実行するには、**セルの左側にある ▶（再生ボタン）をクリック**します。

💡 キーボード操作でも実行できます（慣れてくると便利です）：
- **Shift + Enter**：セルを実行して、次のセルに移動
- **Ctrl + Enter**（または Mac では Cmd + Enter）：セルを実行して、そのままそのセルにとどまる

<br>

---

## 🖥 ランタイムとは？仮想パソコンのイメージ

Google Colabでは、ノートブックを実行すると**クラウド上に一時的な仮想パソコン（＝ランタイム）**が用意されます。

### 💻 パソコン（＝ランタイム）は毎回使い捨て

- ノートブックを開くと、**そのときだけ使える「仮想パソコン」**が起動します。
- プログラムを実行すると、**変数やデータはこのパソコンの中に一時的に保存**されます。
- しかし、パソコンを閉じたり、しばらく使わずにいると、**このパソコンは削除されて中身もすべて消えます**。

### 🧾 出力結果（print文など）はプリントアウトされた紙

- 実行した結果はノートに残って見えますが、**これは「紙に印刷された結果」だけ**。
- 中身（変数やファイル）はすでにパソコンの中から消えていて、**再利用はできません**。

<br>

---

## 💾 データの保存について

Colabでは、**変数の中身などはランタイムが終了すると消えてしまいます**。  
保存できるのは、**ファイルに書き出したデータだけ**です。

- 必要なデータは **ファイルとして保存してからダウンロード**するか、  
- **Googleドライブを「外付けUSBメモリ」**のように接続して、そこに保存することができます。

👉 各変数そのもの（メモリ上の状態）は保存できません。必要なら**ファイルに変換（CSVなど）してから保存**しましょう。

# プログラムが実行される順番
---

## 🧾 基本ルール：上から下に、順番に実行される

プログラムは、**上から順番に1行ずつ**読まれて、実行されていきます。

```python
x = 5
y = 3
print(x + y)
```

このときの実行の流れはこうなります。

1. `x = 5` → x という変数に 5 を入れる
2. `y = 3` → y という変数に 3 を入れる
3. `print(x + y)` → x と y の合計（8）を表示する

<br>

---

## 🧪 Google Colab のルール：**セルごとに実行できる**

Google Colab では、コードを「セル」というかたまりに分けて書きます。

そして、それぞれのセルを**バラバラに実行することができます**。

```python
# セル1
x = 1  # xに1を代入する
```

```python
# セル2
x += 1  # xの数値を1増やす
```

```python
# セル3
print(x)  # xの中身を出力（表示）する
```

- セル1 を**先に実行**しておけば、セル2 と セル3 はうまく動きます（x = 1 が記憶されている）
- **セル1 をスキップ**するとエラーになります（x の中身がまだ決まっていない）
- セル2 はスキップすることが可能です。セル3 では、スキップしない場合は 2 が、スキップした場合は 1 が出力されます。
- セル2 を複数回実行することが可能です。セル2 を実行した回数だけ、xの値は大きくなります。
- セル2 の後で セル1 を実行すると値が 1 に初期化されます。

**基本的に、上から順番に1つずつセルが実行することが前提です。**
セルを飛ばしたり、順番を入れ替えた場合、実行した場合、意図しないエラーや、誤った処理が行われる可能性が高いです。

<br>

- 値がなんだかおかしい

1. 画面上部のメニューから「**ランタイム（Runtime）**」をクリック

この処理は、例えるならば「コンピューターの再起動」ではなく、「コンピューターの破壊と再購入」です。不適切な処理に起因するエラーは全て解決しますが、これまでのデータは全て消えます。

<br>

## 変数とは？


変数とは、**データに名前をつけて記憶するためのしくみ**です。プログラムの中で数値や文字列などのデータを扱うとき、値をそのまま繰り返し記述するのではなく、一度変数に代入しておくことで、再利用や管理がしやすくなります。

たとえば、次のように書いた場合：

```python
x = 10
```

このコードは、「x という名前の変数に 10 という値を代入する」ことを意味します。以後、`x` を使うことで `10` と同じ意味を持たせることができます。

<br>


---
## 変数名のルール

変数名には以下のような命名規則があります。

- 英文字（a〜z、A〜Z）、数字（0〜9）、アンダースコア（_）を使用できる
- **ただし最初の文字は数字にできない**
- ハイフン（-）やスペース、記号（@, %, # など）は使用できない
- **Pythonの予約語（`for`, `if`, `class` など）は変数名として使えない**

例：

```python
valid_name = 123    # 有効
_underscore = "OK"  # 有効だが非推奨（通常の変数は最初をアンダーバーにしない）
Value_A = 0         # 有効だが非推奨（変数は全て小文字、定数は全て大文字）
ValueA = 0          # 有効だが非推奨（変数は全て小文字、定数は全て大文字）
VALUE_A = 0         # 有効（定数は全て大文字でもいい）
2nd_value = 5       # 無効（数字で始まっている）
my-name = "NG"      # 無効（ハイフン使用）
my name = "NG"      # 無効（スペース使用）
```

変数名は、その中身が何を表しているのかを他人（あるいは将来の自分）が理解しやすいようにつけるのが望ましいです。

## `=` は数学の「=」ではない

Pythonにおける `=` は、**「等しい」ではなく「代入する」**という意味を持ちます。

たとえば、以下のコード：

```python
x = x + 1
```

は、数学的には矛盾しているように見えます（x は常に自分自身より1大きくはなれない）が、プログラムにおいては、

- 右辺の `x + 1` を計算し、
- その結果を左辺の `x` に「代入する」

という手順になります。つまり、これは「x の値を1増やす」という操作です（`x += 1`とも書けます）。

<br>

Pythonで「等しいことを比較」したい場合は、`==` を使用する必要があります。

```python
x = 5      # 代入
x == 5     # x が 5 と等しいかどうかを判定
```

<br>

````python
x = 5         # 代入
a = (x == 5)  # x が 5 と等しいかどうかを判定し、その結果を代入
print(a)
````

## 変数の存在意義とは？


プログラムにおいて変数を使う目的は、**値を一時的に記憶し、繰り返し利用したり、意味を明確にしたりすること**にあります。変数を使わずに、直接数値や文字を記述することも可能ですが、**可読性・柔軟性・再利用性**の点で大きな違いが生じます。

<br>

---

### 例：可読性の向上

#### （変数なしの場合）

```python
# 半径5の円の面積を求める
print(3.14 * 5 * 5)
```

#### （変数ありの場合）

```python
radius = 5
pi = 3.14
area = pi * radius * radius
print(area)
```

このように変数を使うことで、「何の値を使っているのか」が明確になり、式の意味を理解しやすくなります。また、値を変更したいときにも、変数を書き換えるだけで済みます。

<br>

----

### 例：柔軟性・再利用性の向上

#### （変数なしの場合）

```python
print("Aさん：こんにちは、Bさん。今日はいい天気ですね。")
print("Bさん：こんにちは、Aさん。ほんとに気持ちがいいですね。")
print("Aさん：Bさんはこのあと予定ありますか？")
print("Bさん：Aさんと一緒にカフェに行こうかと思ってました。")
print("Aさん：それはいいですね、Bさん。行きましょう。")
```

#### （変数ありの場合）

```python
name_a = "Aさん"

```

このように変数を使うことで、**名前を一括で変更したり、複数の場所で繰り返し使用する**ことができます。

## 変数に入れられる「モノ」: 基本的なデータ型

Python の変数には、あらゆるモノ（オブジェクト）を代入することができます。代入できるオブジェクトはいくつかの種類に分けることができます。**基本的な種類（データ型）** は以下です。

| データ型       | 例                    | 説明                             |
| -------------- | --------------------- | -------------------------------- |
| 整数（int）    | `10`, `-3`            | 小数点のない数値                 |
| 小数（float）  | `3.14`, `0.0`         | 小数点を含む数値                 |
| 文字列（str）  | `"hello"`, `'abc123'` | 文字の並び（数値ではない）       |
| 真偽値（bool） | `True`, `False`       | 「はい」か「いいえ」などの論理値 |

他にもリスト、辞書、セット、変数、クラスなど様々なオブジェクトを代入できます。リスト、辞書、セットに関しては後述します。

<br>

### 🔍 「型」の存在意義は？

変数には、数値・文字列・真偽値など、さまざまな種類のデータを入れることができます。しかし、それぞれ**扱い方が異なるため、同じ箱（変数）では管理できません**。

たとえば：

* 文字列は長さがまちまちで、特別な処理も多いため、大きくて柔軟な箱が必要になります。
* 真偽値（`True` / `False`）は「1か0か」の情報しか持たないため、小さくて単純な箱で十分です。

このように、**データの種類ごとに必要な「箱の大きさや機能」が異なる**のです。

Python では、変数の型に応じて適切な箱（メモリ領域）を自動で選び、**無駄なく、効率よくプログラムを動かす**ことができるようになっています。

型を適切に区別することで、処理の安全性が高まり、メモリの無駄を防ぐこともできます。

<br>

## 型の重要性（数値と文字列の比較）


**数値と文字列はまったく別の型であり、見た目が似ていても動作が異なります。**

似た見た目（たとえば `123` と `"123"`）でも、**それが「計算できる数」なのか「ただの文字」なのか**によって、プログラムの動作はまったく異なります。

`"123" + 1` のように、異なる型を混ぜて使おうとするとエラーになります。したがって、**データ型の違いは初心者でも意識しておく必要があります**。

<br>

### 例1：`123 + 123` と `"123" + "123"` の違い

```python
print(123 + 123)        # → 246（数値の加算）
print("123" + "123")    # → "123123"（文字列の連結）
```

- 数値（int や float）は、加算・減算などの計算に使えます。
- 文字列（str）は、**文字としての並び**として扱われ、足し算は「つなげる」意味になります。

<br>

### 例2：`12 * 5` と `"12" * 5` の違い

```python
print(12 * 5)         # → 60（掛け算）
print("12" * 5)       # → "1212121212"（繰り返し）
```

- `12 * 5` は、通常の掛け算です。
- `"12" * 5` は、「文字列 '12' を5回繰り返す」という意味になります。

<br>

---

### 🔁 数値と文字列の型変換

Python では、**数値と文字列を自動的に混ぜて使うことはできません**。異なる型を組み合わせて使う場合は、**明示的に型を変換（型変換）**する必要があります。

#### 🔸 数値 → 文字列 に変換

```python
x = 123
str_x = str(x)  # "123" という文字列に変換
```

#### 🔸 文字列 → 数値 に変換

```python
s = "123"
int_s = int(s)      # 整数に変換（例: 123）
float_s = float(s)  # 小数に変換（例: 123.0）
```

## 変数に入れられる「モノ」: リスト



リストとは、**複数の値（データ）をまとめて1つの変数に入れる**ことができる入れ物です。
たとえば、果物の名前を1つずつ別の変数に入れるのではなく、1つのリストにまとめることができます。

```python
fruits = ["apple", "banana", "orange"]
```

このように、**カンマで区切って角かっこ（`[]`）で囲む**ことで、複数の値を1つのリストにまとめることができます。

<br>

---

### ✅ 今は「覚える」より「存在を知る」<font color="red">【重要！】</font>

初心者にとって重要なのは、以下のような感覚を持つことです：

* 「**リスト**という**ものがある**」
* 「**複数のデータをまとめて扱う**ために使われる」
* 「あとで中身を見たり、増やしたり、順番を変えたりできる」

つまり、**「知らないと検索もできない」状態を脱すること**が第一の目的です。

<br>

### ✅ 必要になったら都度調べる、で十分 <font color="red">【重要！】</font>

* 実務でも上級者でも、細かい構文をすべて暗記している人は少数派です。
* 概念を理解し、**都度検索・ドキュメント参照で正確な使い方にたどり着ける**ことが重要です。

<br>

<details>
<summary>🔍 解説を見る</summary>

---

### 🔍 リストの中身を取り出す（インデックス）

リストの中の値（要素）には、**番号（インデックス）を使ってアクセス**します。番号は **0から始まる** ことに注意してください。

```python
print(fruits[0])  # → "apple"
print(fruits[2])  # → "orange"
```

#### ❗ 注意：

* **存在しない番号**を指定するとエラーになります。

```python
print(fruits[10])  # IndexError（エラー）
```

* **マイナスの番号**を使うと、後ろから数えることができます。

```python
print(fruits[-1])  # → "orange"（最後の要素）
print(fruits[-2])  # → "banana"（後ろから2番目）
```

<br>

---

### ✂ スライス（部分的に取り出す）

リストから、**連続した複数の要素を取り出す**こともできます。これを「スライス」と呼びます。

```python
print(fruits[0:2])  # → ["apple", "banana"]
```

* `0:2` は「0番目から2番目の**手前まで**」を意味します。

省略やマイナスも使えます：

```python
print(fruits[:2])    # → ["apple", "banana"]
print(fruits[1:])    # → ["banana", "orange"]
print(fruits[-2:])   # → ["banana", "orange"]
print(fruits[:-1])   # → ["apple", "banana"]
```

#### ❗ 注意：

* 存在しない範囲を指定してもエラーにはならず、**空のリスト `[]` が返されます**。

```python
print(fruits[5:10])  # → []
```

<br>

---

### ➕ 要素を追加する

リストの最後に新しい要素を追加するには、`append()` を使います。

```python
fruits.append("grape")
print(fruits)  # → ["apple", "banana", "orange", "grape"]
```

#### 💡補足：

* **同じ値でも何回でも追加できます**。

```python
fruits.append("banana")
print(fruits)  # → ["apple", "banana", "orange", "grape", "banana"]
```

* **最初は空のリストを作って、後から追加していく**という使い方もよくあります。

```python
numbers = []
numbers.append(1)
numbers.append(2)
print(numbers)  # → [1, 2]
```

<br>

---

### ➖ 要素を削除する

特定の要素を削除するには、`remove()` または `del` を使います。

```python
fruits.remove("banana")  # 指定した値を削除（最初の1つだけ）
print(fruits)

del fruits[1]  # 指定した位置を削除
print(fruits)
```

<br>

---

### 🔃 要素の並び替え（ソート）

リストの中身を並び替えるには、`sort()` を使います。

```python
numbers = [5, 1, 3, 4]
numbers.sort()
print(numbers)  # → [1, 3, 4, 5]
```

逆順にしたいときは `reverse=True` を使います：

```python
numbers.sort(reverse=True)
print(numbers)  # → [5, 4, 3, 1]
```

<br>

---

### 💡 他にもこんな機能があります（詳細は省略）

* `insert()`：指定位置への要素追加
* `pop()`：指定位置の要素を取り出して削除
* `index()`：特定の値が最初に現れる位置を取得
* `count()`：特定の値の出現回数をカウント
* `copy()`：リストのコピーを作成
* `extend()`：他のリストを結合して拡張
* リスト内包表記（list comprehension）
* 多次元リスト（リストの中にリスト）

</details>

## 変数に入れられる「モノ」：辞書

辞書とは、**「名前」と「値」をセットで保存する**データ構造です。
Python では、「キー（key）」と「値（value）」をペアにして記録します。

```python
person = {"name": "Taro", "age": 20}
```

この場合：

* `"name"` がキー、`"Taro"` がその値
* `"age"` がキー、`20` がその値

まるで「項目名つきのメモ帳」のように、**それぞれの値に意味がある名前（キー）をつけて保存**できます。

<br>

---

### 今は「覚える」より「存在を知る」 <font color="red">【重要！】</font>

辞書もリストと同様に、**今は「こういうものがある」という認識だけで十分です**。具体的な使い方は、必要になったときに繰り返し見直せば問題ありません。

<br>

<details>
<summary>🔍 解説を見る</summary>

---

### 🔍 値の取り出し（キーでアクセス）

リストは番号でアクセスしますが、辞書では**キーを使ってアクセス**します。

```python
print(person["name"])  # → "Taro"
print(person["age"])   # → 20
```

#### ❗ 注意：

* 存在しないキーを指定するとエラーになります。

```python
print(person["height"])  # KeyError（エラー）
```

* 存在しないかもしれないキーに安全にアクセスしたい場合は、`get()` を使います。キーが存在しない場合、通常は`None`が返りますが、帰り値を指定することもできます。

```python
print(person.get("height"))        # → None
print(person.get("height", "不明")) # → "不明"
```

<br>

---

### 🆕 要素の追加・変更

辞書に新しい項目を追加したり、既存の値を上書きしたりできます。空の辞書（`new_dict = {}`）を作り、項目を追加していくことも多いです。

```python
person["height"] = 170  # 新しいキーと値を追加
person["age"] = 21      # 既存のキーの値を更新
```

<br>

---

### ➖ 要素の削除

特定のキーとその値を削除したい場合は `del` を使います。

```python
del person["height"]
```

存在しないキーを削除しようとするとエラーになるので注意してください。

<br>

---

### 🔄 キーや値の一覧を取得する

辞書に入っている情報を確認したいときは以下のようにします：

```python
print(person.keys())    # → dict_keys(['name', 'age'])
print(person.values())  # → dict_values(['Taro', 21])
print(person.items())   # → dict_items([('name', 'Taro'), ('age', 21)])
```

* `keys()`：すべてのキー
* `values()`：すべての値
* `items()`：すべてのキーと値のペア

<br>

---

### 🧠 他にもこんな機能があります（詳細は省略）

* `update()`：複数のキーと値を一度に追加・更新
* `pop()`：指定したキーの値を取り出して削除
* `clear()`：すべての内容を削除
* ネストされた辞書（辞書の中に辞書）
* 辞書内包表記（dictionary comprehension）

リストの中に辞書を入れたり、辞書の値にリストを入れたりすることもできます（辞書のkeyにはリストは指定できない（タプルならばOK））。

</details>

<br>

## 変数に入れられる「モノ」：セット

**セット（set）**は、**重複のない値の集まり**を表すデータ構造です。
数学の「集合」と同じような働きをします。

```python
colors = {"red", "blue", "green"}
```

* `{"..."}` の形で書きます（中カッコでカンマ区切り）。
* **同じ値が2回以上入っても、自動的に1つにまとめられます。**
* **順番は保持されません**（表示の順番は変わることがあります）。
* **集合演算を使って共通部分や差を調べることができます。**

<br>

---

### 今は「覚える」より「存在を知る」 <font color="red">【重要！】</font>


今の段階では、**セットとは「重複しない集まり」だということと、順番がないという特徴だけ覚えておけば十分です。**
実際に使う場面が出てきたときに、詳しい使い方を調べれば問題ありません。

<br>

<details>
<summary>🔍 解説を見る</summary>

<br>

---

### 🔁 重複しないという特徴

```python
colors = {"red", "blue", "red"}
print(colors)  # → {"red", "blue"}
```

"red" が2回入っていても、1つにまとめられて保持されます。

<br>

---

### 🆕 要素の追加

`add()` を使って1つの要素を追加できます。

```python
colors.add("yellow")
print(colors)  # → {"red", "blue", "yellow", ...}
```

* 同じ値を何回追加しても、1つしか保持されません。
* 空のセットを作り、値を追加していくことも多いです。
* **辞書と同じ中かっこ `{}` を使う**ので混同しやすいですが、空のセットを作るときは注意が必要です：

```python
empty_set = set()  # 空のセット
empty_dict = {}    # これは空の辞書
```

<br>

---

### ➖ 要素の削除

指定した要素を削除するには `remove()` や `discard()` を使います。

```python
colors.remove("blue")
```

#### ❗注意：

* `remove()` は、**存在しない要素を削除しようとするとエラーになります**
* `discard()` は、**存在しない場合でもエラーになりません**

```python
colors.discard("pink")  # エラーにならずスルーされる
```

<br>

---

### 🔄 集合演算（リストや辞書にはない特徴）

セットは、**集合演算**（数学の集合と同じ）に特に強みがあります：

```python
a = {1, 2, 3}
b = {3, 4, 5}
```

| 操作  | 説明          | 書き方     | 結果             |                   |
| --- | ----------- | ------- | -------------- | ----------------- |
| 和集合 | 両方にある全ての要素  | `a \| b`            | `{1, 2, 3, 4, 5}` |
| 積集合 | 両方に共通する要素   | `a & b` | `{3}`          |                   |
| 差集合 | aにあってbにないもの | `a - b` | `{1, 2}`       |                   |
| 対称差 | 一方にだけある要素   | `a ^ b` | `{1, 2, 4, 5}` |                   |

<br>

---

### ❗ よくある注意点

* セットは順序がなく、**インデックスで要素を取り出すことはできません**。

  * つまり `colors[0]` のような書き方はできません（エラーになります）
* **辞書と同じ中かっこ `{}` を使う**ので混同しやすいですが、空のセットを作るときは注意が必要です：

```python
empty_set = set()  # 空のセット
empty_dict = {}    # これは空の辞書
```

<br>

---

### 🧠 他にもこんな機能があります（詳細は省略）

* `union()`, `intersection()`, `difference()`：集合演算の関数版
* `issubset()`, `issuperset()`：部分集合かどうかの判定
* `copy()`：セットのコピー
* `clear()`：全要素の削除
* 集合内包表記（set comprehension）

</details>

<br>

## 関数とは？

関数とは、**あらかじめ決められた処理を実行してくれる命令のセット**です。
たとえば、`print()` を使えば、画面に文字を表示してくれます。

関数を使うには、**関数の名前のあとにカッコ `()` をつけて書くだけ**です。

```python
print("こんにちは")
```

**関数を自作することもできますが、まずは既存関数の使用方を学びましょう。**

<br>

---

## 🗣️ 関数は「命令文」

関数を使うときは、「〇〇してね！」と命令しているイメージです。

| 関数        | 意味      | 例                  |
| --------- | ------- | ------------------ |
| `print()` | 表示してね   | `print("こんにちは")`   |
| `len()`   | 長さを教えてね | `len("apple")` → 5 |
| `int()`   | 数に変えてね  | `int("123")` → 123 |

<br>



---

## 🔸 関数の呼び出し方

関数には複数の呼び出し方があり、よく混乱するポイントとなります。

<br>

### ✅ 単独の関数（組み込み関数）

```python
print("こんにちは")
len("apple")
```

* **最も基本的な呼び出し方**
* `print`, `len`, `int`, `str`, など、Pythonに最初から入っている関数
* 変数を箱だとすれば、箱の外にある加工機（Python本体に存在）の呼び出し

<br>

### ✅ オブジェクトについた関数（メソッド）

```python
fruits = ["apple", "banana"]
fruits.append("orange")
```

* **変数名.関数名(...)** という形
* `append()` はリスト専用の関数（このような関数を「メソッド」と呼びます）
* **リストや文字列など、特定の型だけが持つ専用の機能**
* 変数を箱だとすれば、箱に付属する加工機の呼び出し

<br>


### ✅ ライブラリの関数（モジュールの関数）

```python
import numpy as np
np.sum([1, 2, 3])
```

* **ライブラリ名.関数名(...)**
* NumPy など外部ライブラリの関数を使うときによく見る形
* `np` は `numpy` を短く呼びたいときに付ける**別名（エイリアス）**
* 変数を箱だとすれば、箱の外にある加工機（追加した拡張部屋に存在）の呼び出し

<br>



---

## 🧳 カッコの中に「材料」を渡す（引数）

関数には、「何を使って処理するか」を**カッコの中に渡します**。これを**引数（ひきすう）**といいます。

```python
print("おはよう")  # 「おはよう」を表示
len("banana")      # "banana" の長さ → 6
```

関数によって、**引数が1つ必要なもの、複数必要なもの、いらないもの**があります。

<br>

### ✅ 値だけを並べる（位置引数）

```python
print("Hello")
range(1, 11, 2)
```

* 引数が1つの場合は単純に値を括弧内に入れる
* 引数が複数ある場合は**カンマで区切って順番に並べる**
* 関数ごとに「何番目に何を入れるか」が決まっている

<br>

### ✅ キーワード=値 の形で指定（キーワード引数）

```python
sorted(numbers, reverse=True)
```

* **どんな意味の値かを明示する**方法
* 位置引数とキーワード引数の両方が存在することが多い
* キーワード引数は必ず**位置引数の後に書く**
* 引数の順番に縛られず、読みやすく・間違いにくくなる

<br>



---

## 🔁 一部の関数は「結果」を返す

関数の中には、**何かを返す（答えをくれる）タイプ**のものもあります。

```python
length = len("apple")
print(length)  # → 5
```

このように、関数を使って得られた結果を**変数に入れてあとで使う**こともできます。

<br>

### ✅ 値を返さない関数（何かをするだけ）

```python
print("Hello")
mylist.append("orange")
```

* `print()` や `append()` のように、**何かの操作だけして結果は返さない**関数もある
* 変数に入れても `None`（＝なにもない）しか入らないので注意

```python
result = print("Hello")  # → result は None
```

<br>

### ✅ 複数の値を返す関数

```python
import os
name, ext = os.path.splitext("data.csv")
```

* `return` で複数の値を返すことができる
* その場合、**カンマ区切りで複数の変数に受け取る**

<br>

# 条件分岐



---

## 🔀 条件分岐とは？

**「ある条件を満たすときだけ実行する」「条件によって動きを変える」**というのが「条件分岐」です。
Python では、**`if` 文**を使って実現します。

<br>





---

## ✅ 基本の形（if 文）

```python
x = 10

if x > 5:
    print("x は 5 より大きい")
```

* `if` のあとに **条件**（例：`x > 5`）を書きます
* `:` をつけて、次の行に**字下げ（インデント）**して実行する処理を書きます
* 条件が **True（真）** のときだけ実行されます

<br>



---

## 🔁 よく使う比較・含有チェックの書き方

| 記号       | 意味      | 例                        |
| -------- | ------- | ------------------------ |
| `==`     | 等しい     | `x == 5`                 |
| `!=`     | 等しくない   | `x != 3`                 |
| `>`      | より大きい   | `x > 2`                  |
| `<`      | より小さい   | `x < 10`                 |
| `>=`     | 以上      | `x >= 7`                 |
| `<=`     | 以下      | `x <= 10`                |
| `in`     | 含まれている  | `"apple" in fruits`      |
| `not in` | 含まれていない | `"banana" not in fruits` |

* `=` は代入、`==` は比較です（完全に違います）。
* `fruits` はリストやセットなどの複数のデータをまとめたもの（例：`["apple", "orange"]`）です。
* `in` / `not in` は 文字列・リスト・辞書（キー）・セット などに使えます。

<br>



---

## 🔄 if 〜 else（どちらかを実行）

```python
x = 3

if x >= 5:
    print("5以上です")
else:
    print("5未満です")
```

* 条件に合うときは `if` のブロック
* 合わないときは `else` のブロックが実行されます

<br>



---

## 🔀 if 〜 elif 〜 else（条件を複数分ける）

```python
score = 85

if score >= 90:
    print("とてもよくできました")
elif score >= 70:
    print("よくできました")
else:
    print("がんばりましょう")
```

* `elif` は「else if」の略で、条件を追加できます
* 上から順に判定されて、**最初に当てはまった1つだけ**が実行されます

<br>


---

## 🔗 複数の条件を組み合わせる（`and`, `or`）

### ✅ `and`：「両方とも正しいときだけ True」

```python
age = 20
is_student = True

if age >= 18 and is_student:
    print("学生の大人です")
```

→ 両方の条件（年齢が18以上、かつ学生）を満たすときにだけ実行されます。

<br>

### ✅ `or`：「どちらかが正しければ True」

```python
score = 92

if score >= 90 or score == 77:
    print("とても優秀です")
```

→ 「90点以上」または「77点ちょうど」なら実行されます。

<br>

### ✅ `not`：「条件を反対にする」

```python
is_logged_in = False

if not is_logged_in:
    print("ログインしてください")
```

→ `is_logged_in` が False のときに実行されます。

<br>


---

## 【重要】📝 インデントの操作方法（深く／浅く）

Python では、`if` 文や `for` 文、関数の中などで**インデント（字下げ）**が必要です。インデントは、**処理のまとまり**を表す重要なルールです。

| 操作        | Windows         | Mac             |
| --------- | --------------- | --------------- |
| 深くする（下げる） | `Tab`           | `Tab`           |
| 浅くする（上げる） | `Shift` + `Tab` | `Shift` + `Tab` |

- 💡 複数行を選択して実行すれば、**まとめてインデント調整**ができます。

<br>


---
## ⚠️ if 文のよくある間違い

- ❌ :（コロン）の付け忘れ
- ❌ =（代入）と ==（比較）を間違える
- ❌ 文字列・数値を混ぜて比較してしまう
- ❌ インデント（字下げ）がない／ずれている
- ❌ 順番が不適切（大きい条件が後）

## 🔁 ループ処理とは？

**ループ（繰り返し）処理**とは、同じ種類の処理を**何度も自動で繰り返す**仕組みのことです。

* 同じことを何度も手で書くのではなく、**1つのルールでまとめて処理できる**。
* データの一覧を順番に処理したいときなどに非常に便利。

<br>

### ✅ 偶数をprintするコード（forを使わない場合）

~~~python
i = 1
if i % 2 == 0:
    print(i)

i = 2
if i % 2 == 0:
    print(i)

i = 3
if i % 2 == 0:
    print(i)

i = 4
if i % 2 == 0:
    print(i)

i = 5
if i % 2 == 0:
    print(i)
~~~

- `i % 2 == 0` は偶数であることを意味します（2で割って余りが0）

<br>

### ✅ 偶数をprintするコード（forを使う場合）

~~~python
for i in range(1, 6):
    if i % 2 == 0:
        print(i)
~~~

- `range(1, 6)` は 1～5 の整数を生成します（※6は含まれません）

1. ループ1: 1行目（`for i in range(1, 6):`）が実行され、`i` に `1` が代入される
2. ループ1: 2行目（`if i % 2 == 0:`）が実行され、`i` は `1` なので、`1 % 2` は `1` → **False** と判定される
3. ループ1: 判定は False なので、3行目（`print(i)`）は **実行されない**
1. ループ2: 1行目のループが継続し、`i` に `2` が代入される
2. ループ2: 2行目が実行され、`2 % 2` は `0` → **True** と判定される
3. ループ2: 判定は True なので、3行目が実行され、**`2` が出力される**
1. ループ3: 次のループで、`i` に `3` が代入される
2. ループ3: 2行目が実行され、`3 % 2` は `1` → **False**
3. ループ3: 判定は False なので、3行目は **実行されない**
1. ループ4: 次のループで、`i` に `4` が代入される
2. ループ4: 2行目が実行され、`4 % 2` は `0` → **True**
3. ループ4: 判定は True なので、3行目が実行され、**`4` が出力される**
1. ループ5: 次のループで、`i` に `5` が代入される
2. ループ5: 2行目が実行され、`5 % 2` は `1` → **False**
3. ループ5: 判定は False なので、3行目は **実行されない**
1. `range(1, 6)` の値はすべて取り出し終えたので、`for` **ループが終了する**
* 最終的な出力は以下のとおり：

<br>

---

🔽🔽🔽 ⬇ **この下のコードを実行してみましょう！** 🔽🔽🔽


````python
print("ループを開始します")

for i in range(1, 6):
    print(f"このループを開始します(i = {i})")
    print(f"{i}の判定を始めます")
    if i % 2 == 0:
        print(f"{i}は偶数です")
    print(f"このループを開始します(i = {i})")

print("全てのループを終了しました")
print(f"注意！iには値が格納されたままです（i = {i}）")
````

## 🔹 Pythonでよく使うループ：`for` 文

Pythonで最もよく使うループは `for` 文です。
以下のように書きます：

```python
for 変数 in データの集まり:
    処理
```

インデント（字下げ）された部分が、**繰り返す処理**になります。

<br>

### ✅ 例1：リストの中身を順番に表示

```python
fruits = ["apple", "banana", "orange"]

for fruit in fruits:
    print(fruit)
```

出力：

```
apple
banana
orange
```

▶ `fruits` の中から1つずつ取り出して、`fruit` に入れ、表示しています。

<br>

### ✅ 例2：数字を繰り返す – `range()` と組み合わせる

```python
for i in range(5):
    print(i)
```

出力：

```
0
1
2
3
4
```

▶ `range(5)` は「0から4まで」の数字を自動で作ってくれる関数です。
　よく「指定した回数だけ繰り返したい」ときに使われます。

はい、以下は **「forの中にforを書く（入れ子・ネスト）」処理の初心者向け解説**です。

<br>

### 🧪 例3：for の中に for を書く（ネストされたループ）

```python
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j)
```

出力：

```
1 1
1 2
1 3
2 1
2 2
2 3
3 1
3 2
3 3
```

## 🔸 ループの中で使える特別なキーワード

| キーワード | 意味                          | 例                                 |
| ---------- | ----------------------------- | ---------------------------------- |
| `break`    | ループを途中でやめる          | 条件を満たしたら終了したい場合など |
| `continue` | その1回をスキップして次へ進む | 特定の条件のときだけ飛ばしたい場合 |

<br>

```python
for i in range(10):
    if i == 3:
        continue
    if i == 5:
        break
    print(i)
```

- `i == 3`のとき、`continue`が実行されるので、以降の処理はスキップされる（printされない）。ループ自体は継続する。
- `i == 5`のとき、`break`が実行されるので、ループが終了し、以降は一切出力されない。

<br>

---

🔽🔽🔽 ⬇ **この下のコードを実行してみましょう！** 🔽🔽🔽


````python
print("ループ開始")
for i in range(10):
    if i == 3:
        continue
    if i == 5:
        break
    print(i)
print("ループ終了")
````

## 🔁 ループ処理のもう1つの方法：`while`

Pythonでは繰り返し処理の方法として、`for` だけでなく `while` も用意されています。

```python
while 条件:
    繰り返したい処理
```

`while` は「**条件が満たされている間は、何度でも同じ処理を繰り返す**」という仕組みです。

```python
i = 0
while i < 5:
    print(i)
    i += 1
```

この例では、`i` が `5` になるまで `print(i)` が繰り返されます。
※ `i += 1` を忘れると、条件がずっと `True` のままになり、無限ループになるので注意！

<br>

### ✅ まずは `for` を使おう

初心者のうちは、`while` よりも `for` を使う方がわかりやすく、安全です。

* **決まった回数だけ繰り返す**
* **リストや範囲などの中身を順番に処理する**

このような場合には、`for` の方がミスも少なく、読みやすいコードになります。

---

### 💡 よく使う応用：`while True` + `break`

それでも `while` がよく使われるパターンがあります。それが次のような書き方です。

```python
while True:
    条件に応じて break で抜ける
```

これは「**いつ終わるか決まっていないけれど、途中で条件を満たせば終わる**」ような処理に使われます。

<br>

#### 例：0が入力されるまで数字を入力させる

```python
while True:
    num = input("数字を入力してください（0で終了）：")
    if num == "0":
        break
    print("あなたが入力したのは", num)
```

- このように、**入力やユーザーとのやり取り**、**一定条件までの試行**などでは `while True:` を使うことが多くなる
- break の位置を間違えると「止まらない無限ループ」になってしまうことがある
    - Colabで無限ループになってしまったときは、「■（停止ボタン）」 を押す
    - それでも止まらないときは、以下の手順を試す（データは消失）
        1. 上部メニューの「ランタイム」をクリック
        2. 「ランタイムを終了」または「ランタイムを再起動」を選択


<br>

---

🔽🔽🔽 ⬇ **この下のコードを実行してみましょう！** 🔽🔽🔽


````python
# 0から10までの整数をランダムに選び続けて、0が出るまで回す「ガチャ」コード
import random  # ランダムな数字を作るための「random」という道具を使えるようにする

count = 0  # 何回ガチャを回したかを数えるための変数を0でスタート

while True:  # 「ずっと繰り返す」命令（止めるまで無限に繰り返す）
    number = random.randint(0, 10)  # 0から10の中からランダムに1つ数字を選ぶ
    print(f"ガチャ結果: {number}")  # 出た数字を画面に表示する
    count += 1  # ガチャを1回回したので、回数を+1する

    if number == 0:  # 出た数字が0だった場合は
        break  # ループ（繰り返し）を終了する

print(f"0が出るまでに回した回数: {count}")  # 最後に回数を表示する
````
