## 💡 Google Colabとは？

Google Colab（コラボ）は、**ブラウザだけでPythonを使える無料のツール**です。  
パソコンにPythonをインストールしなくても、すぐにコードを書いて実行できます。

- Googleのクラウド上で動く「仮想のパソコン」を使います。
- Googleアカウントがあれば、誰でも利用できます。
- コードと説明文をまとめて書けるので、学習や実験ノートにも便利です。

<br>

---

## 🧱 「セル」とは？

Google Colab のノートは、「セル」と呼ばれる**ブロック単位**でできています。

- 📄 **テキストセル**：説明や見出しを書くところ（この文もテキストセル）
- 🧮 **コードセル**：Python の命令を書くところ（ここに `print("Hello")` などを書く）

👇 セルを実行するには、**セルの左側にある ▶（再生ボタン）をクリック**します。

💡 キーボード操作でも実行できます（慣れてくると便利です）：
- **Shift + Enter**：セルを実行して、次のセルに移動
- **Ctrl + Enter**（または Mac では Cmd + Enter）：セルを実行して、そのままそのセルにとどまる

<br>

---

## 🖥 ランタイムとは？仮想パソコンのイメージ

Google Colabでは、ノートブックを実行すると**クラウド上に一時的な仮想パソコン（＝ランタイム）**が用意されます。

### 💻 パソコン（＝ランタイム）は毎回使い捨て

- ノートブックを開くと、**そのときだけ使える「仮想パソコン」**が起動します。
- プログラムを実行すると、**変数やデータはこのパソコンの中に一時的に保存**されます。
- しかし、パソコンを閉じたり、しばらく使わずにいると、**このパソコンは削除されて中身もすべて消えます**。

### 🧾 出力結果（print文など）はプリントアウトされた紙

- 実行した結果はノートに残って見えますが、**これは「紙に印刷された結果」だけ**。
- 中身（変数やファイル）はすでにパソコンの中から消えていて、**再利用はできません**。

<br>

---

## 💾 データの保存について

Colabでは、**変数の中身などはランタイムが終了すると消えてしまいます**。  
保存できるのは、**ファイルに書き出したデータだけ**です。

- 必要なデータは **ファイルとして保存してからダウンロード**するか、  
- **Googleドライブを「外付けUSBメモリ」**のように接続して、そこに保存することができます。

👉 各変数そのもの（メモリ上の状態）は保存できません。必要なら**ファイルに変換（CSVなど）してから保存**しましょう。

# 個別・少数の情報取得（手作業）
---

遺伝子数が少ない場合は、各種データベースの検索機能を使って、遺伝子名を直接入力して情報を取得する方法でも十分に対応できます。
例えば、ヒトのTP53のように、よく知られた遺伝子であれば、検索窓に「TP53」と入力するだけで、詳細な情報ページにアクセスできます。

ただし、データベースごとに採用している遺伝子IDやアノテーションの基準が異なるため、以下の点に注意が必要です。

- データベース間でIDや座標が異なる場合がある（例：NCBIとEnsemblでUTRの位置が微妙に異なる）
- RefSeqとEnsemblの転写産物（アイソフォーム）が完全には一致しないこともある

後述するゲノムワイド処理は、Ensemblのデータを使用することから、本講習ではEnsemblの情報を優先します。

<br>

---

## 🔁 処理の流れ

### 🧭 スタート地点：まずEnsemblで検索

* [https://www.ensembl.org](https://www.ensembl.org) の検索窓に、
  例：`TP53` または `TP53 human` のように**遺伝子名＋種名**を入力
* 対象が見つかれば、Ensembl ID（例：`ENSG00000141510`）を含むページへ移動

<br>

---

## ✅ パターン1：**Ensemblで見つかった場合**

* 構造情報（exon/CDS/UTR）や配列（genomic/mRNA/CDS）が正確かつ統一された形式で取得できる
* 外部リンクから文献・タンパク質情報もたどれる

<br>

### 🔽 取得ステップ：

| 情報の種類                 | 取得方法                                | ページ例・補足                                                                            |
| --------------------- | ----------------------------------- | ---------------------------------------------------------------------------------- |
| 遺伝子の基本情報              | Gene Summaryタブで確認                   | 例：[TP53 gene](https://www.ensembl.org/Homo_sapiens/Gene/Summary?g=ENSG00000141510) |
| ゲノム座標、イントロン含む配列       | Gene → Sequence                     | 領域ごとのFASTA出力可                                                                      |
| 各アイソフォームのmRNA・CDS・UTR | Transcript → Sequence               | 色分け＋FASTAダウンロード可                                                                   |
| タンパク質配列               | Transcript → Protein                | アミノ酸配列表示・保存可                                                                       |
| 🔗 **外部リンクから追加情報を取得** | Geneページ下部の “External references” にて | NCBI Gene / UniProt / InterPro などへリンクあり                                            |

<br>

### 🧭 プロモーター領域（転写開始点上流）の取得方法

1. [Transcriptページ](https://asia.ensembl.org/Homo_sapiens/Transcript/Summary?db=core;g=ENSG00000141510;r=17:7661779-7687546;t=ENST00000269305) を開く（先程のGeneから開けるページ）
1. Export Data メニューで「周辺領域を含めて出力」が可能
1. 「5' Flanking sequence (upstream)」に1000などを指定して、プロモーター配列を含めたFASTAを出力

ターミネーター領域も同様に「3' Flanking sequence (downstream)」を指定する。

<br>

---

## ❌ パターン2：**Ensemblで見つからなかった場合**

* 非モデル生物（マイナーな微生物、環境サンプル、非参照ゲノムなど）
* 発見されて間もない遺伝子や暫定的アノテーション（hypothetical gene）

<br>

### 🔽 取得ステップ：

| 情報の種類             | 取得方法                                                                  | 補足                                                    |
| ----------------- | --------------------------------------------------------------------- | ----------------------------------------------------- |
| 🧬 遺伝子情報          | NCBI Geneページで検索（例：[TP53](https://www.ncbi.nlm.nih.gov/gene/7157)） | 遺伝子名・種名で検索可能。基本情報やRefSeqへのリンクも一覧表示される                 |
| 🧾 mRNA配列（FASTA）  | Geneページ内の「NCBI Reference Sequences」からたどる                        | RefSeq mRNA（例：NM\_000546.6）へのリンクあり、FASTA形式で取得可能       |
| 🧩 CDS・UTRなどの構造情報 | mRNAページで「GenBank形式」に切り替えて確認*                                      | feature欄に CDS / 5'UTR / 3'UTR の位置情報あり。 |

# 個別・少数の情報取得（API活用）
---

少数の遺伝子であっても、**手作業で調べてコピー＆ペーストするよりも、APIを活用して自動的に取得する方が圧倒的に効率的**です。

<br>

---

## 🧠 APIとは？

**API（Application Programming Interface）**とは、「Web経由で情報を取得するための自動アクセス手段」です。今回のような生物データベース（EnsemblやNCBIなど）でも、検索や配列の取得を**Pythonコードから自動で実行**できるようにAPIが提供されています。

<br>

---

## 📌 以下を行います：

指定した**遺伝子記号（例：TP53）**と**生物種（例：homo\_sapiens）**に対して、以下の処理を**自動で実行する関数**を作成し、活用します。

<br>

---

### 🔄 自動で実行される処理：

1. **Ensemblで遺伝子を検索し、基本情報を取得**

2. 指定した**最大数の転写産物**について：

   * mRNA（cDNA）配列
   * 5'UTR、CDS、3'UTR
   * アミノ酸配列（コーディング転写産物の場合）
   * プロモーター領域（転写開始点の上流領域）
   * ターミネーター領域（転写終了点の下流領域）

3. 取得した情報を**表（DataFrame）として表示**

4. **TSVファイルとして保存し、Colab環境ならダウンロードも可能**

<br>

---

## ✅ 関数を使う前に

以下のセルを実行して、関数本体を読み込んでください（1回だけでOKです）。

<br>

🔽🔽🔽 ⬇ **以下のコードを実行してみましょう！** 🔽🔽🔽


````python
import os
import requests
import pandas as pd
from google.colab import files

def get_gene_and_transcripts_info(
    species: str,
    symbol: str,
    promoter_len: int = 1000,
    terminator_len: int = 500,
    max_transcripts: int = 10,
    output_dir: str = "/content",
):
    """
    Get gene and transcript information from Ensembl using REST API.

    Parameters:
        species (str): Species name used in Ensembl (e.g., "homo_sapiens").
        symbol (str): Gene symbol (e.g., "TP53").
        promoter_len (int): Number of bases upstream of transcription start site to include as promoter.
        terminator_len (int): Number of bases downstream of transcription end site to include as terminator.
        max_transcripts (int): Maximum number of transcripts to retrieve.
        output_dir (str): Directory to save output TSV file.

    Returns:
        tuple: (gene lookup JSON object, output file path)
    """

    headers_json = {"Content-Type": "application/json"}
    headers_fasta = {"Content-Type": "text/x-fasta"}

    # Step 1: Retrieve Ensembl Gene ID
    xref_url = f"https://rest.ensembl.org/xrefs/symbol/{species}/{symbol}?"
    xref_res = requests.get(xref_url, headers=headers_json)
    if xref_res.status_code != 200 or not xref_res.json():
        print(f"❌ Ensemblで '{symbol}' が見つかりませんでした。")
        print("🔍 遺伝子名に誤りがある可能性があります。NCBIで確認してください:")
        print(f"https://www.ncbi.nlm.nih.gov/gene/?term={symbol}+{species}")
        return None

    gene_id = xref_res.json()[0]["id"]

    # Step 2: Retrieve gene and transcript data
    lookup_url = f"https://rest.ensembl.org/lookup/id/{gene_id}?expand=1"
    lookup = requests.get(lookup_url, headers=headers_json).json()

    # Step 3: Format basic gene information
    gene_info = {
        "Gene name": lookup.get("display_name"),
        "Ensembl Gene ID": gene_id,
        "Chromosome": lookup.get("seq_region_name"),
        "Start": lookup.get("start"),
        "End": lookup.get("end"),
        "Strand": lookup.get("strand"),
        "Description": lookup.get("description", "").split("[")[0].strip(),
        "Biotype": lookup.get("biotype"),
        "Transcript count": len(lookup.get("Transcript", [])),
        "Canonical Transcript": lookup.get("canonical_transcript")
    }

    header_lines = [f"# {k}: {v}" for k, v in gene_info.items()]
    print(f'Gene name       : {gene_info["Gene name"]}')
    print(f'Gene ID         : {gene_info["Ensembl Gene ID"]}')
    print(f'Transcript count: {gene_info["Transcript count"]}')

    # Step 4: Iterate over transcripts and collect information
    rows = []
    for t in lookup["Transcript"][:max_transcripts]:
        tid = t["id"]
        print(f'Transcript ID   : {tid}')
        tname = t.get("display_name", tid)
        is_coding = "protein_coding" in t.get("biotype", "")

        # Fetch sequence by type (cdna, cds, utr, protein)
        def fetch_seq(seqtype):
            url = f"https://rest.ensembl.org/sequence/id/{tid}?type={seqtype}"
            res = requests.get(url, headers=headers_fasta)
            if res.status_code == 200:
                lines = res.text.strip().split("\n")
                return "".join(lines[1:])  # Remove FASTA header
            return None

        # Determine promoter and terminator coordinates based on strand
        t_start = t.get("start")
        t_end = t.get("end")
        chrom = gene_info["Chromosome"]
        strand = gene_info["Strand"]

        if strand == 1:
            promoter_start = max(1, t_start - promoter_len)
            promoter_end = t_start - 1
            terminator_start = t_end + 1
            terminator_end = t_end + terminator_len
        else:
            promoter_start = t_end + 1
            promoter_end = t_end + promoter_len
            terminator_start = max(1, t_start - terminator_len)
            terminator_end = t_start - 1

        # Fetch sequence for promoter/terminator regions
        def fetch_region(start, end):
            region_url = f"https://rest.ensembl.org/sequence/region/{species}/{chrom}:{start}..{end}:{strand}"
            r = requests.get(region_url, headers=headers_fasta)
            if r.status_code == 200:
                return "".join(r.text.strip().split("\n")[1:])
            return None

        # Fetch mRNA and extract UTR/CDS if coding
        cdna = fetch_seq("cdna")
        if is_coding:
            cds = fetch_seq("cds")
            utr_list = cdna.split(cds)
            if len(utr_list) == 2:
                utr_5 = utr_list[0]
                utr_3 = utr_list[1]
            else:
                utr_5 = None
                utr_3 = None
        else:
            utr_5, cds, utr_3 = None, None, None

        # Assemble row for each transcript
        row = {
            "Gene": gene_info["Gene name"],
            "Transcript ID": tid,
            "Transcript name": tname,
            "Coding": is_coding,
            "mRNA (cDNA)": cdna,
            "5'UTR": utr_5,
            "CDS": cds,
            "3'UTR": utr_3,
            "Protein (aa)": fetch_seq("protein") if is_coding else None,
            f"Promoter ({promoter_len}bp)": fetch_region(promoter_start, promoter_end),
            f"Terminator ({terminator_len}bp)": fetch_region(terminator_start, terminator_end),
        }
        rows.append(row)

    # Step 5: Create DataFrame and save to TSV
    df = pd.DataFrame(rows)
    display(df)

    if output_dir != "":
        os.makedirs(output_dir, exist_ok=True)
    path = os.path.join(output_dir, f"{species}.{symbol}.tsv")
    with open(path, "w") as f:
        f.write("\n".join(header_lines) + "\n")
        df.to_csv(f, sep="\t", index=False)

    print("✅ TSVファイルを作成しました:", path)
    return lookup, path


print("関数の準備が完了しました。")
````

# ゲノムワイドな情報取得
---

少数の遺伝子を対象とする場合は、APIを用いて1つずつ検索・取得する方法が便利です。しかし、**数百〜数万の遺伝子を一括で扱う場合は、毎回APIで取得する方法は推奨されません**。

その理由は以下のとおりです：

* 🔄 **通信回数が非常に多くなり、処理が遅くなる**
* 🚫 **APIにはアクセス制限（レートリミット）があり、途中でブロックされることがある**
* 🌐 **インターネット接続が不安定な場合に処理が失敗しやすい**
* 🔁 **同じ情報を何度も取得するのは非効率**


そのため、ここでは **ゲノム全体の配列情報（FASTA形式）** と **遺伝子アノテーション情報（GFF3形式）** を使って、一括処理を行う方法を紹介します。

<br>

---

## 📂 FASTAファイルとは？

**FASTA形式**は、DNA・RNA・タンパク質などの**塩基配列やアミノ酸配列を記録するための最も基本的なファイル形式**です。

`.fa`, `.fasta` などの拡張子で保存されることが多いです。

<br>

### 🔹 構造

FASTAファイルは、**1つまたは複数の配列を「ヘッダー行」と「配列本体」に分けて記述**します。

```
>sequence_id 説明文など
ATGCGTAGCTAGCTAGGCTAG...
```

* `>`（大なり記号）で始まる行が「ヘッダー行」です。
* 2行目以降は、\*\*配列の文字列（A, T, G, Cなど）\*\*が記載されます。

<br>


### 🔸 補足事項

* `>`に続く最初の文字列（スペースより前）が「**配列ID**」として扱われ、それ以降は**説明文**として記録されます。
  例：

  ```
  >ENST00000456328.2 Transcript of DDX11L1
  ```

  → IDは `ENST00000456328.2`、説明は `Transcript of DDX11L1`

* 配列は**一定の長さごとに改行される**のが一般的です（例えば60〜80文字ごと）。
* 非常に大きなサイズになることがあります（Google Driveの容量圧迫に注意）。
* サイズが大きいため、ダウンロード提供される際には**gzip圧縮（`.gz` 拡張子）**された状態になっているのが普通です。その場合は拡張子が `.fa.gz`, `.fasta.gz` となります。
* Windows環境での作成・編集時の注意（改行コード）

  * Windowsでは通常、改行コードが `CRLF`（キャリッジリターン + ラインフィード）になります。
  * しかし、Unix系のツールやColab、Linux環境では `LF`（ラインフィード）を標準とします。解析ツール（Webアプリケーションの内部ツール含む）はLinux環境で実行されていることも多く、**FASTAファイルは `LF` で保存するのが推奨**されます。
  * 多くのテキストエディタ（例：VS Code、Notepad++）では改行コードを指定・変換できます。

## 📄 GFF3ファイルとは？

**GFF3（General Feature Format version 3）\**は、ゲノム上の\**遺伝子や転写産物などの位置や構造情報（アノテーション）を記述する形式**です。

- `.gff`, `.gff3` といった拡張子で保存されます。
- **9列のタブ区切りテキスト**で構成されており、各行が遺伝子やエキソンなどの要素を表します。
- 列名（ヘッダー行）は存在しません。誤って列名をデータに含めてしまうと、解析結果に混入したり、エラーの原因になることがあります。
- GFF3では、`#` で始まる行は コメント行 です。
- データの途中や末尾に 空白行（空行）が含まれていることがあります。
-  attributes列の中身は「URLエンコード（%記法）」されていることがあります。
- `.gff3.gz` のように**gzip圧縮されていることが多い**ので、読み込み時は注意しましょう。

```
seqid  source  type       start  end    score  strand  phase  attributes
```


```
chr1   ensembl gene       11869  14409  .      +       .      ID=gene:ENSG00000223972;Name=DDX11L1;biotype=transcribed_unprocessed_pseudogene
chr1   ensembl transcript 11869  14409  .      +       .      ID=transcript:ENST00000456328;Parent=gene:ENSG00000223972
chr1   ensembl exon       11869  12227  .      +       .      ID=exon:ENSE00002234944;Parent=transcript:ENST00000456328
```
| 列番号 | 列名           | 内容例・説明                                                                                                             |
| --- | ------------ | ------------------------------------------------------------------------------------------------------------------ |
| 1   | `seqid`      | 対象配列の名前（例: `chr1`, `NC_000001.11`）                                                                                 |
| 2   | `source`     | アノテーションの由来（例: `Ensembl`, `RefSeq`）                                                                                 |
| 3   | `type`       | フィーチャーの種類（例: `gene`, `mRNA`, `CDS`, `exon`）                                                                        |
| 4   | `start`      | 開始位置（**1-based**、対象領域の先頭）                                                                                          |
| 5   | `end`        | 終了位置（**1-based**、対象領域の末尾）                                                                                          |
| 6   | `score`      | スコア情報（数値または `.`。未定義時は `.`）                                                                                         |
| 7   | `strand`     | 転写方向：`+` または `-`                                                                                                   |
| 8   | `phase`      | コーディング開始のコドン位置（`0`, `1`, `2`）。CDS以外は`.`                                                                            |
| 9   | `attributes` | 属性情報（`;` 区切りの `key=value` ペア）。例：`ID=transcript:ENST00000335137;Parent=gene:ENSG00000141510;biotype=protein_coding` |

<br>

## 🔸 属性カラムの `ID` と `Parent` の関係が重要

GFF3では、**各要素がツリー状（親子関係）で構造化**されています：

* `ID=` で定義されたエントリが「親」となり、
* `Parent=` により、その親の配下に「子」がぶら下がる形になります。

このような構造により、**遺伝子 → 転写産物 → エキソン/CDS** の階層構造を簡潔に表現できます。

<br>

## 🔹 主な feature（フィーチャー）の種類

| type                        | 意味・説明                  |
| --------------------------- | ---------------------- |
| `gene`                      | 遺伝子全体。最も上位の構造。         |
| `mRNA`                      | 転写産物（または `transcript`） |
| `exon`                      | エキソン。スプライシングで残る領域。     |
| `CDS`                       | 翻訳される領域（コーディング配列）      |
| `five_prime_UTR`            | 5'UTR。翻訳されない前方領域。      |
| `three_prime_UTR`           | 3'UTR。翻訳されない後方領域。      |
| `start_codon`, `stop_codon` | 翻訳の開始/終了位置。            |

<br>

## 🌲 親子関係の典型的な例

```
gene
└── mRNA / transcript
    ├── exon
    ├── CDS
    ├── five_prime_UTR
    └── three_prime_UTR
```

このように、**構造の階層が明確に定義されているため、解析や可視化の際に非常に扱いやすくなっています。**

<br>

## 📝 FASTAとGFF3の入手先に関する注意

ゲノム配列（FASTA）とアノテーション情報（GFF3）を使って遺伝子情報を解析する場合、**両者の整合性が非常に重要**です。以下の点に注意してください。

<br>

### ✅ 染色体名の「ゆらぎ」に注意

FASTAとGFF3で **同じ染色体を指していても名前の表記が異なる** ことがあります。例えば、染色体1番は次のように表記されることがあります：

* `1`
* `chr1`
* `Chr1`
* `chr01`

このような差異があると、**FASTAとGFF3を対応させて使うことができなくなります**。特にFASTAとGFF3を別々のサイトからダウンロードした場合に起こりやすいので、**セットで提供されているものを使う**のが安全です。

<br>

### 🔬 データ提供元による特徴と注意点

| データ提供元                        | 特徴                                        | 注意点                                             |
| ----------------------------- | ----------------------------------------- | ----------------------------------------------- |
| **Ensembl**                   | 多くの種を統一形式で管理。FASTAとGFF3の整合性が高く、定期的に更新される。 | データは様々な元から集めた統合版であり、特定種に特化した最新情報は含まれていない可能性も。   |
| **種特異的DB**（例: FlyBase、TAIRなど） | 特定の生物種に特化しており、最新の遺伝子情報やWebアプリが使える。        | GFF3の記述形式が非標準であることがあり、ツールで読めないことも。形式の整備が不十分な例も。 |

* Ensemblのファイルが形式がよく整っており、機械的に読み込み処理するのに適しています。
* 初学者は、**EnsemblからFASTAとGFF3をセットでダウンロードする**のがおすすめです。

## GFF3ファイルの読み込みと遺伝子情報の取得

このセクションでは、**GFF3ファイルを読み込み、各遺伝子ごとに以下の情報をまとめて表形式で出力する方法**を紹介します。

* 遺伝子名やID
* 染色体やゲノム上の位置（start, end）
* ストランド（転写方向）
* 遺伝子の長さ
* 各遺伝子に属する転写産物の数
* 重複を除いたエキソンの数
* biotype などの付加情報（attributes列から）

<br>

### 使用するライブラリ

ここでは、**GFF3ファイルを効率的に読み込むために `gffutils` ライブラリ**を使います。
gffutils は、GFF3ファイルをデータベース（SQLite）に変換し、柔軟な検索や親子関係のたどりができる便利なツールです。

<br>

### 処理の概要

1. GFF3ファイルのパスを指定します。
3. `.db` ファイルがなければ新たにデータベースを作成し、ある場合はそれを再利用します。
4. 各 gene フィーチャーについて以下を取得します：

   * 染色体・ストランド・位置などの基本情報
   * attributes列にある情報（gene\_nameやbiotypeなど）
   * 直接の子要素である転写産物（mRNAやncRNAなど）の数
   * さらにその子要素である exon の重複を排除した数
5. 上記を pandas の DataFrame にまとめ、画面に表示します。
6. 最後に TSVファイルに保存し、Colab でダウンロードできるようにします。

<br>

🔽🔽🔽 ⬇ **以下のコードを実行してみましょう！** 🔽🔽🔽

````python
# gffutils がインストールされていない場合のみインストールする
try:
    import gffutils
except ImportError:
    !pip install gffutils
    import gffutils

import pandas as pd
import os

# GFF3ファイルのパス（必要に応じて変更）
gff_file = "/content/Caenorhabditis_elegans.WBcel235.61.chromosome.I.gff3.gz"
db_fn = gff_file.removesuffix(".gz") + ".db"
output_path = gff_file.removesuffix(".gz").removesuffix(".gff3") + ".gene_info.tsv"

# GFFデータベースの作成
if not os.path.exists(db_fn):
    print("GFFデータベースの作成中...（しばらく時間がかかります）")
    # 一般的にはkeep_order=True, disable_infer_transcripts=False, merge_strategy="merge"
    # 今回は高速化のためkeep_order=False, disable_infer_transcripts=True, merge_strategy="create_unique"
    db = gffutils.create_db(
        gff_file, db_fn, force=True, keep_order=False, disable_infer_transcripts=True,
        merge_strategy="create_unique", sort_attribute_values=True
    )
    # GFFデータベースの保存
    files.download(db_fn)
else:
    db = gffutils.FeatureDB(db_fn)

# geneごとの情報を抽出、attributeキーも収集
attribute_keys = set()
rows = []

transcript_types = [
    "transcript", "mRNA", "ncRNA", "lnc_RNA", "miRNA", "tRNA",
    "rRNA", "snRNA", "snoRNA", "scRNA", "antisense_RNA",
    "pseudogenic_transcript", "processed_transcript"
]

for gene in db.features_of_type("gene"):
    gene_id = gene.id
    chrom = gene.chrom
    start = gene.start
    end = gene.end
    strand = gene.strand
    length = end - start + 1

    attr_dict = {k: v[0] for k, v in gene.attributes.items()}
    attribute_keys.update(attr_dict.keys())

    # 重複のないexonをカウント
    transcript_count = 0
    exon_set = set()
    for transcript in db.children(gene, featuretype=transcript_types, level=1):
        transcript_count += 1
        for exon in db.children(transcript, featuretype="exon"):
            exon_set.add((exon.start, exon.end))

    row = {
        "Name": None,
        "gene_id": None,
        "Chromosome": chrom,
        "Start": start,
        "End": end,
        "Strand": strand,
        "Length": length,
        "Transcript Count": transcript_count,
        "Exon Count": len(exon_set),
    }
    row.update(attr_dict)
    rows.append(row)

# 列を整えてDataFrame作成
main_cols = ["Name", "gene_id", "Chromosome", "Start", "End", "Strand", "Length", "Transcript Count", "Exon Count"]
attribute_keys.remove("Name")
attribute_keys.remove("gene_id")
attribute_cols = sorted(attribute_keys)
gene_df = pd.DataFrame(rows)[main_cols + attribute_cols]

# 表示
display(gene_df)

# ファイルに保存
with open(output_path, "w") as f:
    gene_df.to_csv(f, sep="\t", index=False)

# ファイルをダウンロード
files.download(output_path)
````

## FASTAファイルの読み込みと配列計算

このコードは、**GFF3ファイルとFASTAファイルを用いて、遺伝子と転写産物ごとの構造情報と塩基配列を取得・整理する**処理を行います。Colabなどの環境で、**選択した遺伝子群（ここではランダムに100遺伝子）について、mRNAや各配列領域（UTR、CDS、プロモーター、ターミネーター、タンパク質配列）を抽出**し、最終的に表形式で出力・ダウンロードします。

<br>

### 📦 主な入力ファイル

* `.gff3.gz`：遺伝子構造情報ファイル（Ensembl GFF3形式）
* `.fa.gz`：ゲノム配列ファイル（gzip圧縮されたFASTA形式）

<br>

### 処理の概要

1. GFF3ファイルとFASTAファイルのパスを指定します。

2. `.db` ファイル（gffutils用）がなければ新たに作成し、ある場合はそれを再利用します。

3. GFF3ファイルからgeneフィーチャーをランダムに100個選びます（全遺伝子を対象にすることも可能です）。

4. 各 gene について、以下を取得します：

   * gene\_id、gene\_name、染色体名、ストランドなどの基本情報
   * 直接の子要素である転写産物（mRNAやncRNAなど）
   * 各転写産物について、exonの数
   * 5'UTR、CDS、3'UTR の塩基配列
   * CDSから翻訳されたアミノ酸配列（Protein）
   * 転写開始点 upstream のプロモーター領域、および転写終了点 downstream のターミネーター領域

5. 上記の情報を `pandas` の DataFrame にまとめ、画面に表示します。

6. 最後に TSVファイルとして保存し、Colabでダウンロードできるようにします。

<br>

🔽🔽🔽 ⬇ **以下のコードを実行してみましょう！** 🔽🔽🔽

````python
# gffutils がインストールされていない場合のみインストールする
try:
    import gffutils
except ImportError:
    !pip install gffutils
    import gffutils

# biopython がインストールされていない場合のみインストールする
try:
    from Bio import SeqIO
    from Bio.Seq import Seq
except ImportError:
    !pip install biopython
    from Bio import SeqIO
    from Bio.Seq import Seq

import pandas as pd
import random
import gzip
import os

# ▼ 入力ファイルパス（必要に応じて変更してください）
gff_file = "/content/Caenorhabditis_elegans.WBcel235.61.chromosome.I.gff3.gz"
fasta_file = "/content/Caenorhabditis_elegans.WBcel235.dna.chromosome.I.fa.gz"
db_fn = gff_file.removesuffix(".gz") + ".db"
output_path = gff_file.removesuffix(".gz").removesuffix(".gff3") + ".rna_seq.tsv"

# 取得する配列長を指定
promoter_len = 100
terminator_len = 100

# GFFデータベースの作成
if not os.path.exists(db_fn):
    print("GFFデータベースの作成中...（しばらく時間がかかります）")
    # 一般的にはkeep_order=True, disable_infer_transcripts=False, merge_strategy="merge"
    # 今回は高速化のためkeep_order=False, disable_infer_transcripts=True, merge_strategy="create_unique"
    db = gffutils.create_db(
        gff_file, db_fn, force=True, keep_order=False, disable_infer_transcripts=True,
        merge_strategy="create_unique", sort_attribute_values=True
    )
    # GFFデータベースの保存
    files.download(db_fn)
else:
    db = gffutils.FeatureDB(db_fn)

# 任意の100遺伝子をランダム抽出
sample_genes = random.sample(list(db.features_of_type("gene")), 100)

# 全遺伝子を対象とする場合は以下
# sample_genes = list(db.features_of_type("gene"))

# ゲノム配列の読み込み（辞書形式）
with gzip.open(fasta_file, "rt") as handle:
    genome = {record.id: record.seq for record in SeqIO.parse(handle, "fasta")}

# 転写産物のタイプ（mRNAやncRNAなど）
transcript_types = [
    "transcript", "mRNA", "ncRNA", "lnc_RNA", "miRNA", "tRNA",
    "rRNA", "snRNA", "snoRNA", "scRNA", "antisense_RNA",
    "pseudogenic_transcript", "processed_transcript"
]

# 各遺伝子・転写産物に対して情報を収集
attribute_keys = set()
rows = []

for gene in sample_genes:
    gene_id = gene.attributes.get("gene_id", [""])[0]
    gene_name = gene.attributes.get("Name", [""])[0]
    chrom = gene.chrom
    strand = gene.strand

    for tx in db.children(gene, featuretype=transcript_types, level=1):
        tx_id = tx.id
        transcript_id = tx.attributes.get("transcript_id", [tx_id])[0]
        tx_name = tx.attributes.get("Name", [tx_id])[0]
        biotype = tx.attributes.get("biotype", [""])[0]
        attr_dict = {k: v[0] for k, v in tx.attributes.items()}
        attribute_keys.update(attr_dict.keys())

        # エキソン
        exon_count = sum([1 for _ in db.children(tx, featuretype="exon")])

        # 配列取得関数（リバースストランドも対応）
        def extract_sequence(start, end):
            seq = genome[chrom][start:end]
            return seq.reverse_complement() if strand == "-" else seq

        # 部位ごとの連結配列を取得
        def get_concat_seq(featuretype):
            parts = list(db.children(tx, featuretype=featuretype, order_by='start'))
            if parts:
                seq = Seq("".join(str(genome[chrom][f.start-1:f.end]) for f in parts))
                if strand == "-":
                    seq = seq.reverse_complement()
            else:
                seq = None
            return seq

        utr5 = get_concat_seq("five_prime_UTR")
        cds = get_concat_seq("CDS")
        utr3 = get_concat_seq("three_prime_UTR")

        # プロモーター・ターミネーター領域（開始位置・終了位置の調整）
        if strand == "+":
            promoter = extract_sequence(tx.start - promoter_len-1, tx.start - 1)
            terminator = extract_sequence(tx.end, tx.end + terminator_len)
        else:
            promoter = extract_sequence(tx.end, tx.end + promoter_len)
            terminator = extract_sequence(tx.start - terminator_len-1, tx.start - 1)

        # 結果を記録
        row = {
            "gene_id": gene_id,
            "gene_name": gene_name,
            "transcript_id": transcript_id,
            # "trans_name": tx_name,
            # "biotype": biotype,
            "Exon Count": exon_count,
            "5'UTR": str(utr5) if utr5 else None,
            "CDS": str(cds) if cds else None,
            "3'UTR": str(utr3) if utr3 else None,
            "Protein": str(cds.translate()) if cds else None,
            "Promoter": str(promoter),
            "Terminator": str(terminator),
        }
        # row.update(attr_dict)
        rows.append(row)

# 列を整えてDataFrame作成
main_cols = ["gene_id", "gene_name", "trans_id", "trans_name", "biotype", "Exon Count", "5'UTR", "CDS", "3'UTR", "Protein", "Promoter", "Terminator"]
# attribute_cols = sorted(attribute_keys)
# rna_seq_df = pd.DataFrame(rows)[main_cols + attribute_cols]
rna_seq_df = pd.DataFrame(rows)

# 表示
display(rna_seq_df)

# ファイルに保存
with open(output_path, "w") as f:
    rna_seq_df.to_csv(f, sep="\t", index=False)

# ファイルをダウンロード
files.download(output_path)
````

## 表（DataFrame）からFASTAファイルへ変換・保存

<br>

### ✅ 解説

`rna_seq_df` の各行から以下の情報を使って、FASTA形式のファイル（`test_output.fa`）を作成します：

* `transcript_id` 列の値を **FASTAの名前**（`>名前`）として使用します。
* `gene_id` と `gene_name` 列の値を **FASTAの説明文**として、スペース区切りで追加します（例：`>transcript_id gene_id gene_name`）。
* `5'UTR` 列にある配列を **FASTAの塩基配列**として出力します。

なお、**`5'UTR` の値が `None` または空文字列（長さ0）である場合はスキップ**します。

配列は自動的に **60文字ごとに改行**され、FASTA形式として整形されます（`SeqIO.write` による自動処理）。

<br>

🔽🔽🔽 ⬇ **以下のコードを実行してみましょう！** 🔽🔽🔽

````python
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO

records = []

for _, row in rna_seq_df.iterrows():
    seq = row.get("5'UTR", None)
    if not seq or len(seq) == 0:
        continue

    # 各レコードを作成
    record = SeqRecord(
        Seq(seq),
        id=row.get("transcript_id", ""),
        description=f"{row.get('gene_id', '')} {row.get('gene_name', '')}"
    )
    records.append(record)

# 出力
output_path = "test_output.fa"
SeqIO.write(records, output_path, "fasta")

# ファイルをダウンロード
files.download(output_path)
````
