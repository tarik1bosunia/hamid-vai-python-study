# 🔰 Google Colabの使い方（超入門）
---

このノートブックでは、Python の学習を始める前に、Google Colab の基本的な使い方を説明します。  
パソコンに何もインストールせずに使える便利なツールですが、いくつかの仕組みを知っておくと安心です。

<br>

---

## 💡 Google Colabとは？

Google Colab（コラボ）は、**ブラウザだけでPythonを使える無料のツール**です。  
パソコンにPythonをインストールしなくても、すぐにコードを書いて実行できます。

- Googleのクラウド上で動く「仮想のパソコン」を使います。
- Googleアカウントがあれば、誰でも利用できます。
- コードと説明文をまとめて書けるので、学習や実験ノートにも便利です。

<br>

---

## 🧱 「セル」とは？

Google Colab のノートは、「セル」と呼ばれる**ブロック単位**でできています。

- 📄 **テキストセル**：説明や見出しを書くところ（この文もテキストセル）
- 🧮 **コードセル**：Python の命令を書くところ（ここに `print("Hello")` などを書く）

👇 セルを実行するには、**セルの左側にある ▶（再生ボタン）をクリック**します。

💡 キーボード操作でも実行できます（慣れてくると便利です）：
- **Shift + Enter**：セルを実行して、次のセルに移動
- **Ctrl + Enter**（または Mac では Cmd + Enter）：セルを実行して、そのままそのセルにとどまる

<br>

---

## 🖥 ランタイムとは？仮想パソコンのイメージ

Google Colabでは、ノートブックを実行すると**クラウド上に一時的な仮想パソコン（＝ランタイム）**が用意されます。

### 💻 パソコン（＝ランタイム）は毎回使い捨て

- ノートブックを開くと、**そのときだけ使える「仮想パソコン」**が起動します。
- プログラムを実行すると、**変数やデータはこのパソコンの中に一時的に保存**されます。
- しかし、パソコンを閉じたり、しばらく使わずにいると、**このパソコンは削除されて中身もすべて消えます**。

### 🧾 出力結果（print文など）はプリントアウトされた紙

- 実行した結果はノートに残って見えますが、**これは「紙に印刷された結果」だけ**。
- 中身（変数やファイル）はすでにパソコンの中から消えていて、**再利用はできません**。

<br>

---

## 💾 データの保存について

Colabでは、**変数の中身などはランタイムが終了すると消えてしまいます**。  
保存できるのは、**ファイルに書き出したデータだけ**です。

- 必要なデータは **ファイルとして保存してからダウンロード**するか、  
- **Googleドライブを「外付けUSBメモリ」**のように接続して、そこに保存することができます。

👉 各変数そのもの（メモリ上の状態）は保存できません。必要なら**ファイルに変換（CSVなど）してから保存**しましょう。

## 💡 コードが見えない？

現在、このセルの下のコードは**非表示**になっています。

興味がある人は、「コードを表示」をクリックして、中身を見てみましょう！

🔽🔽🔽 ⬇ **この下にあるセルをクリックして選択し、「再生ボタン」を押してみよう！ ** 🔽🔽🔽

````python
# @title
import collections

import pandas as pd
from scipy.stats import pearsonr, chi2_contingency
import matplotlib.pyplot as plt


# ------------------------------------------------------------------------------
# 前回の入力があれば退避（text が定義済みかつ有効）
try:
    if 'text' in locals() and len(text) >= 50 and all(ord(c) < 128 for c in text):
        prev_text = text
        prev_counter = collections.Counter(prev_text)
        prev_total = sum(prev_counter.values())
        prev_rel_freq = {k: v / prev_total for k, v in prev_counter.items()}
    else:
        prev_text = None
except:
    prev_text = None

# ------------------------------------------------------------------------------
# ユーザー入力
while True:
    text = input("半角英数字でキーボードを適当に叩いて50文字以上の文字列を入力してください（例: qawsedrftgyhujikol...）:\n")

    # 条件チェック
    if not text:
        print("⚠ 入力が空です。")
    elif len(text) < 50:
        print(f"⚠ 入力が短すぎます（{len(text)}文字）。50文字以上入力してください。")
    elif not all(ord(c) < 128 for c in text):
        print("⚠ 半角文字（ASCII文字）のみを入力してください。全角文字が含まれています。")
    else:
        break  # 条件をすべて満たしたら終了

# ------------------------------------------------------------------------------
# 出現頻度・出現順の計算
counter = collections.Counter(text)
total = sum(counter.values())
rel_freq = {k: v / total for k, v in counter.items()}

first_index = {}
for i, char in enumerate(text):
    if char not in first_index:
        first_index[char] = i

# ------------------------------------------------------------------------------
# DataFrame 作成
all_chars = set(counter.keys())
if prev_text:
    all_chars.update(prev_rel_freq.keys())

df = pd.DataFrame([
    {
        'Character': char,
        'Current Freq': rel_freq.get(char, 0),
        'First Index': first_index.get(char, None),
        **({'Previous Freq': prev_rel_freq.get(char, 0)} if prev_text else {})
    }
    for char in sorted(all_chars)
])
df = df.sort_values(by='Current Freq', ascending=False).reset_index(drop=True)

# 表示
display(df)

# ------------------------------------------------------------------------------
# 棒グラフ
fig, ax = plt.subplots(figsize=(12, 5))
ax.bar(df['Character'], df['Current Freq'], label='Current', alpha=0.7)
if prev_text:
    ax.bar(df['Character'], df['Previous Freq'], label='Previous', alpha=0.5)
ax.set_xlabel('Character')
ax.set_ylabel('Relative Frequency')
ax.set_title('Character Frequency')
ax.grid(True, axis='y', linestyle='--', alpha=0.5)
if prev_text:
    ax.legend()
plt.show()

# ------------------------------------------------------------------------------
# 散布図（出現順 vs 相対頻度）
fig, ax = plt.subplots(figsize=(6, 6))
ax.scatter(df['First Index'], df['Current Freq'], c='blue')
ax.set_xlabel('First Appearance Index')
ax.set_ylabel('Relative Frequency')
ax.set_title('Character Frequency vs. First Appearance')
ax.grid(True, linestyle='--', alpha=0.5)
plt.show()

# 相関係数と無相関の検定（出現順 vs 相対頻度）
valid_corr = df[['First Index', 'Current Freq']].dropna()

if len(valid_corr) >= 2:  # 相関計算には最低2点必要
    r_val, p_val = pearsonr(valid_corr['First Index'], valid_corr['Current Freq'])

    print(f"📈 Pearson correlation: r = {r_val:.3f}, p = {p_val:.4f}")
    if p_val < 0.05:
        print("➡ 出現順と相対頻度の間に有意な相関があります（p < 0.05）")
    else:
        print("➡ 出現順と相対頻度の間に有意な相関は確認されませんでした（p ≥ 0.05）")
else:
    print("⚠ 相関係数の計算に十分なデータがありません。")

# ------------------------------------------------------------------------------
# 前回と今回の相対頻度の比較（散布図）
if prev_text:
    fig, ax = plt.subplots(figsize=(6, 6))

    # 散布図プロット
    ax.scatter(
        df['Previous Freq'],
        df['Current Freq'],
        color='blue'
    )

    # 対角線（同じ頻度ならこの上に乗る）
    ax.plot(
        [0, max(df['Previous Freq'].max(), df['Current Freq'].max())],
        [0, max(df['Previous Freq'].max(), df['Current Freq'].max())],
        linestyle='--', color='gray', linewidth=1
    )

    ax.set_xlabel('Previous Relative Frequency')
    ax.set_ylabel('Current Relative Frequency')
    ax.set_title('Current vs Previous Character Frequency')
    ax.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    # 各文字ごとの出現数（前回＋今回）
    combined_counts = {
        c: prev_counter.get(c, 0) + counter.get(c, 0)
        for c in all_chars
    }

    # しきい値未満の文字を「Other」に統合（例: 合計5未満）
    threshold = 5
    filtered_chars = [c for c in all_chars if combined_counts[c] >= threshold]
    other_chars = [c for c in all_chars if combined_counts[c] < threshold]

    # カウント再構成（統合含む）
    def count_with_other(counter_dict):
        counts = [counter_dict.get(c, 0) for c in filtered_chars]
        other_count = sum(counter_dict.get(c, 0) for c in other_chars)
        return counts + [other_count]

    # 前回と今回の出現数ベクトルを作成
    prev_counts = count_with_other(prev_counter)
    curr_counts = count_with_other(counter)

    # 2行×n列の列連表（contingency table）
    contingency_table = [prev_counts, curr_counts]

    # カイ二乗検定
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # 結果表示
    print(f"📊 Chi-squared test: χ² = {chi2:.3f}, p = {p:.4f}, dof = {dof}")
    if p < 0.05:
        print("➡ 前回と今回の出現頻度に有意な差があります（p < 0.05）")
    else:
        print("➡ 前回と今回の出現頻度に有意な差は確認されませんでした（p ≥ 0.05）")
````

````python
import matplotlib.pyplot as plt
import numpy as np
import random

# ↓ 日本語を使いたい場合は、以下の2行のコメントを外してください
# !pip install japanize-matplotlib
# import japanize_matplotlib

# ------------------------------------------------------------------------------
# 波形データの準備
x = np.linspace(0, 10, 100)  # xの値を生成
y1 = np.sin(x)  # 基本的な正弦波
y2 = np.cos(x)  # コサイン波
y3 = np.sin(2 * x)  # 周期が短いサイン波（← 改造ポイント: 2 を大きくすると細かく、小さくするとゆったり）
y4 = np.exp(-0.3 * x) * np.sin(3 * x)  # 減衰しながら振動する波（← 改造ポイント: -0.3 や 3 を調整してみましょう）

# ------------------------------------------------------------------------------
# 星データの準備
n_stars = 20  # 星の数（← 改造ポイント: 数値を変更してみましょう）
star_x = np.random.uniform(0, 10, n_stars)
star_y = np.random.uniform(-2, 2, n_stars)
star_sizes = np.random.randint(100, 300, n_stars)  # サイズ: 100〜300の整数でランダム（← 改造ポイント: 数値を変更してみましょう）
star_colors = np.random.choice(['gold', 'orange', 'pink', 'skyblue'], n_stars)  # 色: 色リストからランダムに選ぶ

# ------------------------------------------------------------------------------
# 図の作成
fig, ax = plt.subplots(figsize=(10, 6))  # ← 改造ポイント: 図のサイズ（inch単位）を調整できます

# ------------------------------------------------------------------------------
# 線をプロット
# 改造ポイント①: 必要ない線は先頭に「#」をつけてコメントアウトしてみましょう
# 改造ポイント②: color='blue' → 他の色に変えてみましょう（例: 'black', 'orange', 'cyan', 'magenta'）
# 改造ポイント③: linestyle='-' → 線の種類を変えてみましょう（例: '--', '-.', ':'）
ax.plot(x, y1, label='sin(x)', color='blue', linestyle='-')
ax.plot(x, y2, label='cos(x)', color='green', linestyle='--')
ax.plot(x, y3, label='sin(2x)', color='red', linestyle='-.')
ax.plot(x, y4, label='damped wave', color='purple', linestyle=':')

# ------------------------------------------------------------------------------
# ランダムな位置・色に★マークを散りばめる
# 改造ポイント①: もしこの描画が不要なら、この行を「#」でコメントアウトして実行してみましょう
# 改造ポイント②: marker='*' → 他の形に変えてみましょう（例: '*', 'o', '^', 's', 'D', 'X'
ax.scatter(star_x, star_y, s=star_sizes, c=star_colors, marker='*', alpha=0.8, edgecolors='black')

# ------------------------------------------------------------------------------
# 軸の調整
# 改造ポイント①: 調整が不要なら、「#」でコメントアウトして実行してみましょう
# 改造ポイント②: 数値を変えてみましょう
ax.set_xlim(0, 10)           # x軸の範囲
ax.set_ylim(-2, 2)           # y軸の範囲
ax.set_xticks(np.arange(0, 11, 2))  # x軸ラベルの間隔 (最小値, 最大値+α, 間隔)
ax.set_yticks(np.arange(-2, 2.1, 0.5))  # y軸ラベルの間隔 (最小値, 最大値+α, 間隔)

# ------------------------------------------------------------------------------
# タイトル・ラベル
# 改造ポイント①: 不要なら、「#」でコメントアウトして実行してみましょう
# 改造ポイント②: 文字を変えてみましょう
ax.set_title('A Graph with Many Features', fontsize=16)
ax.set_xlabel('Value of x')
ax.set_ylabel('Value of y')

# ------------------------------------------------------------------------------
# 凡例ラベル
# ax.legendというキーワードで調べて、以下に凡例を追加してみましょう

# ------------------------------------------------------------------------------
# その他装飾
# 改造ポイント①: 不要なら、「#」でコメントアウトして実行してみましょう
# 改造ポイント②: 設定値を変えてみよう
ax.axhline(0, color='black', linewidth=0.5, linestyle='--')  # 横軸基準線
ax.grid(True, linestyle='--', alpha=0.5)

# ------------------------------------------------------------------------------
# 文字を適当な位置に追加
# 改造ポイント①: 不要なら、「#」でコメントアウトして実行してみましょう
# 改造ポイント②: 設定値を変えてみましょう
ax.text(2, 1.5, 'Python is awesome!', fontsize=12, color='gray', alpha=0.6)
ax.text(7, -1.5, 'Just a sample text', fontsize=12, color='gray', alpha=0.6)

plt.show()
````
