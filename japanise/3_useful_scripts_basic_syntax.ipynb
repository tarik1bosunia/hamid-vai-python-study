## 💡 Google Colabとは？

Google Colab（コラボ）は、**ブラウザだけでPythonを使える無料のツール**です。  
パソコンにPythonをインストールしなくても、すぐにコードを書いて実行できます。

- Googleのクラウド上で動く「仮想のパソコン」を使います。
- Googleアカウントがあれば、誰でも利用できます。
- コードと説明文をまとめて書けるので、学習や実験ノートにも便利です。

<br>

---

## 🧱 「セル」とは？

Google Colab のノートは、「セル」と呼ばれる**ブロック単位**でできています。

- 📄 **テキストセル**：説明や見出しを書くところ（この文もテキストセル）
- 🧮 **コードセル**：Python の命令を書くところ（ここに `print("Hello")` などを書く）

👇 セルを実行するには、**セルの左側にある ▶（再生ボタン）をクリック**します。

💡 キーボード操作でも実行できます（慣れてくると便利です）：
- **Shift + Enter**：セルを実行して、次のセルに移動
- **Ctrl + Enter**（または Mac では Cmd + Enter）：セルを実行して、そのままそのセルにとどまる

<br>

---

## 🖥 ランタイムとは？仮想パソコンのイメージ

Google Colabでは、ノートブックを実行すると**クラウド上に一時的な仮想パソコン（＝ランタイム）**が用意されます。

### 💻 パソコン（＝ランタイム）は毎回使い捨て

- ノートブックを開くと、**そのときだけ使える「仮想パソコン」**が起動します。
- プログラムを実行すると、**変数やデータはこのパソコンの中に一時的に保存**されます。
- しかし、パソコンを閉じたり、しばらく使わずにいると、**このパソコンは削除されて中身もすべて消えます**。

### 🧾 出力結果（print文など）はプリントアウトされた紙

- 実行した結果はノートに残って見えますが、**これは「紙に印刷された結果」だけ**。
- 中身（変数やファイル）はすでにパソコンの中から消えていて、**再利用はできません**。

<br>

---

## 💾 データの保存について

Colabでは、**変数の中身などはランタイムが終了すると消えてしまいます**。  
保存できるのは、**ファイルに書き出したデータだけ**です。

- 必要なデータは **ファイルとして保存してからダウンロード**するか、  
- **Googleドライブを「外付けUSBメモリ」**のように接続して、そこに保存することができます。

👉 各変数そのもの（メモリ上の状態）は保存できません。必要なら**ファイルに変換（CSVなど）してから保存**しましょう。

# コメント選抜プログラム
---

このプログラムは、**英語のサスペンス作品に対するコメント一覧**を自動的に分類してくれるツールです。
内容を分析し、以下のようにコメントを振り分けます：

<br>

## 🎯 プログラムの目的

1. **ポジティブな感想**を含むコメントを集める

   → 作品を褒めているコメントを見つけやすくする

2. **ネタバレなど注意が必要なコメント**を見つけ出す

   → 「犯人」や「ヤス」、「spoiler」などを含むコメントを事前に警告できる

<br>

## 🔁 処理の流れ（ステップごとに解説）

1. **コメントの一覧を用意**

   → サンプルとして英語の感想10件をリストで定義

2. **キーワードリストを準備**
   - ポジティブな単語（例："great", "brilliant" など）
   - ネタバレ関連の危険ワード（例："Yasu", "spoiler", "culprit"）

3. **各コメントに対して次の2つを判定**：

    - **① ポジティブコメントか？**
        - 一定以上の長さがあるか？（例：40文字以上）
        - ポジティブな単語のどれかが含まれているか？

    → 両方を満たしたら「ポジティブコメント」として保存

    - **② 要注意コメントか？**
        - 危険ワードが何個含まれているか数える
        - 2つ以上含まれていたら「要注意」として保存

4. **最後に結果を表示**
   　- 条件を満たしたコメントだけを見やすく一覧表示

<br>

## 🔬 応用アイデア

- アンケートの分析
- テスト回答の分析
- 文章の高度な検索（1文ずつ分割して処理）

<br>

以下はChatGPTに「基本構文のみで書く便利プログラム」を考えていますと前置きをした状態で、以下をお願いした結果**そのままです。**

あるサスペンス英語に関するコメントのリストがあるとして、ポジティブなキーワード集を作っておき、それらを含むコメントだけ集めましょう。ある程度以上長いコメントだけを対象にしましょう。さらに「犯人」、「ヤス」、「ネタバレ」のキーワードのうち2つ以上を含むコメントも別に集めましょう。


<br>

---

🔽🔽🔽 ⬇ **この下のコードを実行してみましょう！** 🔽🔽🔽


```python
# コメント一覧（英語のサスペンスに対するコメント）
comments = [
    "The plot twist was absolutely brilliant and unexpected!",
    "I loved the ending, especially how the culprit was finally revealed.",
    "Yasu is definitely the murderer. Total spoiler!",
    "This is boring.",
    "Great suspense and strong characters. Excellent acting too!",
    "Spoiler alert: Yasu is the culprit. What a twist!",
    "Nice one.",
    "The cinematography was stunning. So intense and emotional.",
    "Honestly, the story was too predictable.",
    "Can't believe it was Yasu all along! Such a clever reveal. Spoiler for sure."
]

# ポジティブキーワード
positive_keywords = ["brilliant", "great", "excellent", "intense", "emotional", "loved", "stunning", "clever"]

# 注意ワード（犯人・ヤス・ネタバレ）
danger_keywords = ["culprit", "Yasu", "spoiler"]

# 長さのしきい値（文字数）
min_length = 40

# 結果格納用
positive_results = []
danger_results = []

for comment in comments:
    # 条件①：ポジティブで、ある程度長い
    if len(comment) >= min_length and any(word in comment.lower() for word in positive_keywords):
        positive_results.append(comment)

    # 条件②：注意ワードが2個以上含まれる
    count = sum(1 for word in danger_keywords if word.lower() in comment.lower())
    if count >= 2:
        danger_results.append(comment)

# 結果表示
print("✅ ポジティブなコメント（長め）:")
for c in positive_results:
    print("-", c)

print("\n⚠️ 要注意コメント（2つ以上の危険ワード）:")
for c in danger_results:
    print("-", c)
```

## コード理解チャレンジ

上記のコードはそのままでも動きますが、初心者が理解するには少々難しい部分もあります。そこで、ChatGPTに解説を加えてもらいつつ、より簡単な表現で書き直してもらいました。

何度かの調整が必要でしたが、得られたコードはほぼChatGPTの出力結果のツギハギだけです。

<br>

---

🔽🔽🔽 ⬇ **この下のコードを「理解」してみましょう！** 🔽🔽🔽


```python
# コメント一覧（英語のサスペンスに対するコメント）
comments = [
    "The plot twist was absolutely brilliant and unexpected!",
    "I loved the ending, especially how the culprit was finally revealed.",
    "Yasu is definitely the murderer. Total spoiler!",
    "This is boring.",
    "Great suspense and strong characters. Excellent acting too!",
    "Spoiler alert: Yasu is the culprit. What a twist!",
    "Nice one.",
    "The cinematography was stunning. So intense and emotional.",
    "Honestly, the story was too predictable.",
    "Can't believe it was Yasu all along! Such a clever reveal. Spoiler for sure."
]

# ポジティブキーワード
positive_keywords = ["brilliant", "great", "excellent", "intense", "emotional", "loved", "stunning", "clever"]

# 注意ワード（犯人・ヤス・ネタバレ）
danger_keywords = ["culprit", "Yasu", "spoiler"]

# 長さのしきい値（文字数）
min_length = 40

# 結果格納用
positive_results = []
danger_results = []

# 各コメントに対して1つずつチェックしていきます
for comment in comments:

    # ------------------------------
    # 条件①：ポジティブで、ある程度長い
    # ------------------------------

    # コメントが長さの条件を満たしているか？
    if len(comment) >= min_length:
        found = False  # ポジティブな単語が見つかったかどうかを記録する変数

        # ポジティブ単語を1つずつ調べる
        for word in positive_keywords:
            # コメントの中にその単語が含まれているか？
            if word in comment.lower():
                found = True  # 見つかったらTrueにする
                break         # 1個見つかれば十分なのでループを止める

        # 見つかっていれば結果に追加
        if found:
            positive_results.append(comment)

    # ------------------------------
    # 条件②：危険ワードを2つ以上含む
    # ------------------------------

    # このコメントに含まれる「危険ワード（ネタバレなど）」の数を数えるための変数を0で初期化します
    count = 0

    # danger_keywords に入っている単語（"culprit", "Yasu", "spoiler"）を1つずつ取り出します
    for word in danger_keywords:

        # 小文字にしたキーワードが、小文字にしたコメント文の中に含まれているか調べます
        # 例: "spoiler" in "spoiler alert: Yasu is the culprit." → True
        if word.lower() in comment.lower():

            # 含まれていれば、カウントを1つ増やします
            count += 1

    # 危険ワードが2つ以上含まれているコメントは「要注意」とみなしてリストに追加します
    if count >= 2:
        danger_results.append(comment)


# 結果表示
print("✅ ポジティブなコメント（長め）:")
for c in positive_results:
    print("-", c)

print("\n⚠️ 要注意コメント（2つ以上の危険ワード）:")
for c in danger_results:
    print("-", c)
```

# 採点結果の可視化（合否で色分け）
---

## 🎯 プログラムの目的

あるクラスの生徒のテスト点数を棒グラフにし、

- **60点以上は青色（合格）**
- **60点未満は赤色（不合格）**

   で表示します。

<br>

以下もChatGPTの出力そのままです。この程度のコードであれば特に修正は必要ないですね。

<br>

---

🔽🔽🔽 ⬇ **この下のコードを「理解」してみましょう！** 🔽🔽🔽


```python
# 🔰 必要なライブラリを読み込みます（matplotlib：グラフ描画用）
import matplotlib.pyplot as plt

# 🧑‍🎓 生徒の名前と点数をリストで定義します
# 名前と点数を1組にした「タプル」のリストです
scores = [
    ("Alice", 85),
    ("Bob", 55),
    ("Charlie", 78),
    ("David", 0),
    ("Eva", 48),
    ("Frank", 90),
    ("Grace", 58)
]

# 📝 名前と点数をそれぞれ別のリストに分けます
# 🟥 合否に応じた色も別リストに追加します
names = []     # 生徒の名前
values = []    # 得点
colors = []    # 棒の色（条件分岐で決定）

# 🔁 各生徒に対して繰り返し処理
for name, score in scores:
    names.append(name)       # 名前を追加
    values.append(score)     # 得点を追加

    # 🎨 合否によって棒の色を決める（条件分岐）
    if score >= 60:
        colors.append("skyblue")  # 合格：水色（青系）
    else:
        colors.append("salmon")   # 不合格：サーモンピンク（赤系）

# 📊 グラフの描画
plt.figure(figsize=(10, 5))            # グラフのサイズを指定
plt.bar(names, values, color=colors)   # 棒グラフを描く（色も指定）

# 📎 グラフのタイトルとラベル
plt.title("Test Scores by Student")
plt.xlabel("Student")
plt.ylabel("Score")
plt.ylim(0, 100)   # 点数の範囲を0〜100に固定

# 💬 合格ラインを点線で表示（参考用）
plt.axhline(y=60, color='gray', linestyle='--', label="Passing Line")

# 🔎 凡例を表示（どの線が何かを説明）
plt.legend()

# 🎉 グラフを表示
plt.show()
```
